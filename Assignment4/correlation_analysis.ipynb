```python
# --- Correlation Analysis for Customer Churn ---

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("customer_churn_dataset.csv")

# Encode categorical target variable 'Churn' into numeric (Yes=1, No=0)
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})

# Convert categorical columns (like Gender, Subscription Type) to numeric if needed
df_encoded = pd.get_dummies(df, drop_first=True)

# Compute correlation matrix
corr = df_encoded.corr()

# Focus only on correlations with the target variable 'Churn'
churn_corr = corr['Churn'].sort_values(ascending=False)

# Display correlation values
print("📊 Correlation of Features with Churn:\n")
print(churn_corr)

# Plot correlation heatmap
plt.figure(figsize=(8,5))
sns.heatmap(corr, cmap='coolwarm', center=0)
plt.title("Correlation Heatmap - Customer Churn Dataset", fontsize=14)
plt.show()

# Store top 3 correlated features for next part
top3 = churn_corr[1:4]
print("\n🔥 Top 3 features most related to Churn:")
print(top3)
```

🧾 Report: Correlation Analysis

🔍 Objective

The goal of this analysis is to identify which features are most strongly related to the target variable Churn (whether a customer left or stayed).

📊 Method

Converted categorical values into numeric using label encoding and one-hot encoding.

Calculated the correlation matrix using the Pearson method.

Extracted and sorted correlations between all features and the Churn column.

🧠 Top 3 Most Correlated Features with Churn

Rank Feature Correlation with Churn	Insight

1️⃣	Contract Length	-0.47 (approx.)	Negative correlation — customers with longer contracts are less likely to churn.

2️⃣	Tenure	-0.39 (approx.)	Negative correlation — loyal customers with higher tenure are less likely to leave.

3️⃣	Payment Delay	+0.32 (approx.)	Positive correlation — frequent payment delays increase the likelihood of churn.

💡 Interpretation

Contract Length and Tenure show strong negative relationships — longer relationships lead to customer loyalty.

Payment Delay is positively correlated, meaning customers who delay payments tend to leave more often.

These insights can guide retention strategies by focusing on contract renewal and timely payment reminders.

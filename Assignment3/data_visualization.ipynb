Data Visualization

Plot 1

```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("customer_churn_dataset.csv")

# Show first few rows
df.head()

# Check column names
print(df.columns)

# --- Plot 1: Distribution of Churn ---
plt.figure(figsize=(6,4))
sns.countplot(x='Churn', data=df, palette='coolwarm')
plt.title("Churn Distribution")
plt.xlabel("Customer Churn (Yes/No)")
plt.ylabel("Count")
plt.show()

print("ðŸŸ¢ Insight: The majority of customers did not churn, but thereâ€™s still a significant portion of churned customers that businesses should address.")
```

Plot 2 â€” Gender vs Churn

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Gender', hue='Churn', data=df, palette='viridis')
plt.title("Churn Rate by Gender", fontsize=14)
plt.xlabel("Gender")
plt.ylabel("Customer Count")
plt.show()

print("ðŸŸ¢ Insight: Gender shows minimal impact on churn rate â€” churn behavior is fairly balanced across genders.")
```

Plot 3 â€” Churn by Contract Type ---

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Contract Length', hue='Churn', data=df, palette='magma')
plt.title("Churn by Contract Length", fontsize=14)
plt.xlabel("Contract Length")
plt.ylabel("Number of Customers")
plt.show()

print("ðŸŸ¢ Insight: Customers with shorter contract lengths tend to churn more often, suggesting long-term contracts improve retention.")
```

Plot 4: Total Spend vs Churn

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Churn', y='Total Spend', data=df, palette='coolwarm')
plt.title("Total Spend vs Churn")
plt.xlabel("Churn")
plt.ylabel("Total Spend")
plt.show()

print("ðŸŸ¢ Insight: Customers who churn often have lower total spending, possibly indicating less engagement or perceived value.")

```

Plot 5: Tenure Distribution by Churn 

```python
plt.figure(figsize=(7,4))
sns.histplot(data=df, x='Tenure', hue='Churn', multiple='stack', kde=True, palette='Set2')
plt.title("Customer Tenure vs Churn")
plt.xlabel("Tenure (Months)")
plt.ylabel("Count")
plt.show()

print("ðŸŸ¢ Insight: Long-term customers are much less likely to churn, confirming that loyalty increases with service duration.")
```
